```python
# calculate medians for churned and retained
```

```python
wave_df.groupby('label').median(numeric_only = True)
```

```python
# churned users averaged more drives in the month than retained users
# retained users use the app on more days than churned
# churned users frve ~200 km more and 2.5 hours more than retained users
```

```python
# median km per drive for retained vs. churned
```

```python
wave_df['km_per_drive'] = wave_df['driven_km_drives'] / wave_df['drives']
median_km_drive = wave_df.groupby('label').median(numeric_only = True) [['km_per_drive']]
median_km_drive
```

```python
# median for retained is 1 km more per drive than churned
```

```python
# calculate km per driving days
```

```python
wave_df['km_per_driving_day'] = wave_df['driven_km_drives'] / wave_df['driving_days']
median_km_driving_day = wave_df.groupby('label').median(numeric_only=True) [['km_per_driving_day']]
median_km_driving_day
```

```python
# churned users drove signifcantly longer distances than retained users
# calculate median number of drives per driving day
```

```python
wave_df['drives_per_driving_day'] = wave_df['drives'] / wave_df['driving_days']

median_drives_per_driving_day = wave_df.groupby('label').median(numeric_only = True) [['drives_per_driving_day']]
median_drives_per_driving_day
```

```python
# churned users have significantly more drives per day than retained users
# need to review more information about these users to determine what leads to the increased drives. 
# This information could have significance in why they churn
```
